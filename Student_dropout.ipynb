```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Student= pd.read_csv("/content/student_dropout_dataset_v3.csv")
#Student.head(25)

# Bo verileri doldurma
Student['Family_Income'] = Student['Family_Income'].fillna(Student['Family_Income'].median())
Student['Stress_Index'] = Student['Stress_Index'].fillna(Student['Stress_Index'].median())

#Is Haritas ,hangi zellik dropout'u etkiliyor?
plt.figure(figsize=(10,6))
sns.heatmap(Student.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='RdYlGn')
plt.title("Correlation Analysis")
plt.show()

#GPA ve Dropout likisi
plt.figure(figsize=(8,5))
sns.boxplot(x='Dropout', y='CGPA', data=Student)
plt.title("Impact of GPA on Dropout")
plt.show()

#Student.isnull().sum()

#Student.info()

#Student.columns

#mod medyan avg gibi deerleri renkli bir tabloda gsterir
Student.describe().style.background_gradient(cmap="Spectral_r")
```
